```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'data latih' : [0.2],
    'random forest' : [0.9444],
    'dare forest' : [0.9444],
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['data latih'], df['random forest'], marker = ".",linestyle='-', color='blue', label='Random Forest')
plt.plot(df['data latih'], df['dare forest'], marker = ".",linestyle='-', color='blue', label='Random Forest')
plt.show()
```
