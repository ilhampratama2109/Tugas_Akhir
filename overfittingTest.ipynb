```python
import dare
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_validate
```

```python
data = pd.read_csv("cleaned_loan.csv")
data
```

```python
labelEncoder = {}
for col in data.columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    labelEncoder[col] = le
data
```

```python
data = data.drop_duplicates()
data
```

```python
X = data.drop(["Keputusan Akhir"], axis=1)
y = data["Keputusan Akhir"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.85, random_state=42)
```

```python
X_train = np.array(X_train.values.tolist())
```

```python
y_train = np.array(y_train.values.tolist())
```

```python
X_test = np.array(X_test.values.tolist())
```

```python
rf = dare.Forest(n_estimators=100, max_depth=5, k = 5, topd = 0, random_state=42)
```

```python
cv_results = cross_validate(rf, X_train, y_train, cv=10, scoring='accuracy')
```

```python
print('Cross-validation scores:')
print(cv_results['test_score'])
print(f'Mean CV score: {cv_results["test_score"].mean()}')
print(f'Standard deviation of CV scores: {cv_results["test_score"].std()}')
```
